```python
## Problem 5 

# Given
L = 32 # number of points
n = np.linspace(0,31, 32) 

xn = 10*np.cos(np.pi*n/16) # Input signal

## Suppose this signal is passed through a nonlinearity with a clipping interval [−8, 8], 
## resulting in a signal: {-8: n < -8, xn: -8 <= n <= 8, 8: n > 8}

yn = 10*np.cos(np.pi*n/16) # Starts the same as x
yn[yn < -8] = -8 # Clip bottom values
yn[yn > 8] = 8 # Clip top values


## Compute 32-point DFTs of xn and yn:

xk = np.fft.fft(xn, L)
yk = np.fft.fft(yn, L)


## Plot the magnitudes of the 32-point DFTs of x[n] and y[n] (for 0 ≤ n ≤ 31) on two separate
## graphs. Please comment on the result.
# Plot 
with plt.style.context('ggplot'): # Set style to be like R
    
    # Plot clipped signal
#     plt.figure()
#     plt.title('Plot of clipped signal')
#     p1 = plt.stem(yn, label = 'y[n]')
#     plt.ylabel('Value') 
#     plt.xlabel('n') 
    
    
    # Plot DFT Magnitudes of x and y
    plt.figure(figsize=(8,4))
    p2 = plt.stem(xk, label = 'DFT{x[n]}')
    plt.ylabel('Magnitude') 
    plt.xlabel('n') 
    plt.title('DFT{x[n]}')
    
    plt.figure(figsize=(8,4))
    p3 = plt.stem(yk, label = 'DFT{y[n]}')
    plt.ylabel('Magnitude') 
    plt.xlabel('n') 
    plt.title('DFT{y[n]}')
    
    
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
    
```
