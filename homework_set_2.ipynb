```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
from scipy import signal # Cant find 'signal' when just scipy
%matplotlib inline
```

```python
## Problem 1 a)

# H(z) = (1/(z^-1 - z^-2 -0.5z^-4))
s1 = signal.lti([1], [0, 1, -1, 0, -0.5])
w, mag, phase = s1.bode()

with plt.style.context('ggplot'): # Set style to be like R
    plt.figure()
    plt.semilogx(w, mag)    # Bode magnitude plot
    plt.xlabel('Frequency')
    plt.ylabel('Magnitude')
    plt.title('1 a) Plot of System Magnitude')
    
    plt.figure()
    plt.semilogx(w, phase)  # Bode phase plot
    plt.xlabel('Frequency')
    plt.ylabel('Phase')
    plt.title('1 a) Plot of System Phase')
    
    plt.show()
```
