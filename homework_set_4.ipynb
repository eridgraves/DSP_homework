```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
import scipy.signal # Cant find 'signal' when just scipy
%matplotlib inline

# Hide warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
## Problem 1 b)
# Plot H(ejw) magnitude and phase using filter coefficients
w, h = scipy.signal.freqz(b=[1.00, 0.00, 1.00], a=[1.00, 0.00, 0.81])

# Plot
with plt.style.context('ggplot'): # Set style to be like R
    fig, ax1 = plt.subplots(figsize=(10,7)) 

    ax1.plot(w,20*np.log10(np.abs(h)),'b') # amplitude
    ax1.set_ylabel('Frequency Response (dB)',color='blue') 
    ax1.set_xlabel('Frequency (radians/sample)') 
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    ax2.plot(w,np.unwrap(np.angle(h)),'--','r') # phase
    ax2.tick_params('y', colors='r') 
    ax2.set_ylabel('Phase Angle (rad)', color='r');
```
