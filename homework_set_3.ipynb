```python
import numpy as np
from matplotlib import pyplot as plt
import scipy
import scipy.signal # Cant find 'signal' when just scipy
import os
import math

%matplotlib inline
```

```python
## Problem 5 a)

# Set up time space
T = 1/3000 # period given
stop_time = 0.001
num_steps = stop_time * T
t = np.arange(0, stop_time, num_steps) # start, stop, step

# Continuous Time
y = np.cos(2000*np.pi*t)

# Sampled Continuous 
sample_time = (int)(stop_time * T)
t_sampled = t[::300] # lis[start : stop : step] , 3000 gives 10 samples over the time scale
y_sampled = np.cos(2000*np.pi*t_sampled)

# Plot
with plt.style.context('ggplot'): # Set style to be like R
    fig, ax1 = plt.subplots(figsize=(5,3)) 
    
    # Plot the continuous signal
    ax1.plot(t, y)
    ax1.set_ylabel('Magnitude (continuous signal)',color='blue') 
    ax1.set_xlabel('Time') 
    
    # Plot the sampled continuous signal on the same axis
    ax2 = ax1.twinx()
    ax2.stem(t_sampled, y_sampled)
    ax2.set_ylabel('Magnitude (discrete signal)',color='r');
    
    
    # Style stuff below
    font = {'family': 'sans',
           'color':  'black',
           'weight': 'normal',
           'size': 16,
           }
    plt.title('5 a) xc(t) = cos(2π(1000)t, T = (1/3000) sec',fontdict = font)
    
    
```
